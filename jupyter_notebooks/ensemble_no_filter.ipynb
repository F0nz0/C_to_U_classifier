```python
# load basic modules
import pandas as pd
```

```python
# iforest model aggregated data
wt_iforest_dfCTaggr_filepath = "/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr//WT.df_CT_predicted_aggregated_iforest_apobec1_pvalues.tsv"
ko_iforest_dfCTaggr_filepath = "/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr//KO.df_CT_predicted_aggregated_iforest_apobec1_pvalues.tsv"
# cnn model aggregated data
wt_cnn_dfCTaggr_filepath = "/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr//WT.df_CT_predicted_aggregated_CNN_apobec1_pvalues.tsv"
ko_cnn_dfCTaggr_filepath = "/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr//KO.df_CT_predicted_aggregated_CNN_apobec1_pvalues.tsv"
```

### WT

```python
wt_iforest_dfCTaggr = pd.read_csv(wt_iforest_dfCTaggr_filepath, index_col=0)
wt_iforest_dfCTaggr
```

```python
wt_cnn_dfCTaggr = pd.read_csv(wt_cnn_dfCTaggr_filepath)
wt_cnn_dfCTaggr
```

```python
# create ensemble predictions for Wt
wt_ensemble = pd.merge(wt_iforest_dfCTaggr, wt_cnn_dfCTaggr, how="inner", on=["region", "position", "strand"])
wt_ensemble
```

```python
# create ensemble predictions for Wt after correction
wt_ensemble_corrected = pd.merge(wt_iforest_dfCTaggr.query("y_hat == 1"), 
                                 wt_cnn_dfCTaggr.query("y_hat == 1"), 
                                 how="inner", on=["region", "position", "strand"])
wt_ensemble_corrected
```
