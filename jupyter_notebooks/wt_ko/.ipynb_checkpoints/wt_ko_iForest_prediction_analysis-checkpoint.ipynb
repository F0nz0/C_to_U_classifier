### WT

```python
# define input paths for iForest prediction on basecalling features
wt_dfCTaggr_filepath = "/lustre/bio_running/C_to_U_editing_minimap2_spliced/wt.model_iForestcc1_pretrained_results/df_CT_predicted_aggregated.tsv"
ref_filepath = "/lustre/bio_running/C_to_U_editing/refs/GRCm39.genome.fa"
```

```python
dfCTaggr = pd.read_table(wt_dfCTaggr_filepath)

plt.figure(figsize=(10,5))
sn.barplot(data=dfCTaggr[["Tfreq_native", "Tfreq_corrected"]])
plt.title("WT - iForest model - CT frequency")
plt.ylabel("Frequency")
sn.despine(offset=10, trim=True)
plt.savefig("/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr/wt_ko/wt.iForest.aggregated.barplot.tiff")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sn.boxplot(data=dfCTaggr[["Tfreq_native", "Tfreq_corrected"]])
plt.yscale("log")
plt.title("WT - iForest model - CT frequency")
plt.ylabel("Frequency")
sn.despine(offset=10, trim=True)
plt.savefig("/lustre/bio_running/C_to_U_editing_minimap2_spliced/src_jupyter_notebooks_multi_thr//wt_ko/wt.iForest.aggregated.boxplot.tiff")
plt.show()
```
